```python
import os
import matplotlib.pyplot as plt
import numpy as np

import numpy
from sklearn.datasets import load_files
from pyvi import ViTokenizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import make_pipeline
```

```python
INPUT = 'data/news_vnexpress'
os.makedirs('image', exist_ok=True)
```

```python
size = 0
for label in os.listdir(INPUT):
    print(f'Loading from dir {label}: {len(os.listdir(os.path.join(INPUT, label)))} files')
    size += len(os.listdir(os.path.join(INPUT, label)))

print(f'Total files loaded: {size}')
```

```python
raw_data = load_files(container_path=INPUT, encoding='utf-8')

for index, label in enumerate(raw_data.target_names, 1):
    print(f'{label}: {index}')

print(f'Total files: {len(raw_data.filenames)}')
```

```python
f = open('data/vietnamese-stopwords-dash.txt', mode='r', encoding='utf-8')
stopWord = f.readlines()

stopWord = [x.strip() for x in stopWord]
print(f'StopWords count: {len(stopWord)}')

count_vectorizer = CountVectorizer(stop_words=stopWord)

model_preprocessing = make_pipeline(
    count_vectorizer,
    TfidfTransformer()
)

data_preprocessed = model_preprocessing.fit_transform(raw_data.data, raw_data.target)

x = data_preprocessed
y = raw_data.target

print(f'Vocabulary size: {len(count_vectorizer.vocabulary_)}')
print(x.shape)
print(y.shape)

```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
model = LinearSVC()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision score: {precision_score(y_test, y_pred, average='micro')}')
print(f'Recall: {recall_score(y_test, y_pred, average='micro')}')
```
